```python
import matplotlib.pyplot as plt
import matplotlib.animation
import time
import random
%matplotlib notebook

xdata = []
ydata = []

axes = plt.gca()
line, = axes.plot(xdata, ydata, 'r-')

counter=0
class Node:
    def __init__(self,val):
        self.val=val

def generateObj():
    a=0
    while True:
        a+=1
        yield Node(a)

def update(i):
    global counter
    counter+=1
    xdata.append(counter)
    ydata.append(i.val)
    line.set_xdata(xdata)
    line.set_ydata(ydata)
    axes.set_xlim(0, counter)
    axes.set_ylim(0, 100)

ani= matplotlib.animation.FuncAnimation(plt.gcf(), update, frames=generateObj(),
                                       interval=100, repeat=False)

plt.show()
```
